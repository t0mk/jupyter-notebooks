```python
import requests
import dateutil.parser
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

url = 'https://api.github.com/repos/ansible/ansible-modules-extras/issues'

issues_dict = {}

try:
    for i in range(1,12):                          
        r = requests.get(url, params={'labels': 'community_review', 'page':i})
        issues_page = r.json()
        if type(issues_page) is not list:
            raise Exception("not a list: %s" % issues_page)
        if len(issues_page) == 0:
            break
        else:
            issues_dict[i] = issues_page

    all_issues = sum(list(issues_dict.values()),[])

    updated_ats = [dateutil.parser.parse(i['updated_at']) for i in all_issues]
except Exception as e:
    print("there was seome problem: %s" % e)
    updated_ats = loaded_issue_dates

n, bins, patches = plt.hist(updated_ats,bins=100)
plt.title("\"updated_at\" of open issues in ansible-modules-extras")
```
